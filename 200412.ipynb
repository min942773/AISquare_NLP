```python
from konlpy.tag import Okt
```

```python
okt = Okt()
```

```python
fread = open('wiki_data.txt', encoding='utf-8')
n = 0
result = []

while True:
    line = fread.readline()
    if not line:
        break
    n = n + 1
    if n % 5000 == 0:
        print('{}번째 while문.'.format(n))
    tokenlist = okt.pos(line, stem=True, norm=True)
    temp = []
    for word in tokenlist:
        if word[1] in ['Noun']:
            temp.append((word[0]))
            
    if temp:
        result.append(temp)
fread.close()
```

```python
print('총 샘플 수 : {}'.format(len(result)))
```

# GloVe(Global Vectors for Word Representation)

- GloVe : 카운트 기반과 예측 기반을 모두 사용하는 방법론
- 카운트 기반의 LSA(Latent Semantic Analysis)와 예측 기반의 Word2Vec의 단점을 보완하는 목적

LSA(잠재 의미 분석) : DTM이나 TF-IDF 행렬과 같이 각 문서에서의 각 단어의 빈도수를 카운트 한 행렬이라는 전체적인 통계정보를 입력으로 받아 차원 축소(Truncated SVD)하여 잠재된 의미를 끌어내는 방법론(모델)

Word2Vec : 실제값과 예측값에 대한 오차를 손실 함수를 통해 줄여나가며 학습하는 예측 기반 방법론(모델)

- LSA 단점 : 단어 의미의 유추 작업에는 성능이 떨어진다. 예) 왕은 남자다. 여왕은 ?
- Word2Vec 단점 : 임베딩 벡터가 윈도우 크기 내에서만 주변 단어를 고려하기 때문에 corpus 전체적인 통계정보를 반영하지 못한다. 

### 윈도우 기반 동시 등장 행렬(Window based Co-occurence Matrix)

- 행과 열을 전체 단어 집합의 단어들로 구성하고, i 단어의 윈도우 크기(Window Size)내에서 k 단어가 등장한 횟수를 i행 k열에 기재한 행렬

### 동시 등장 확률(Co-occurrence Probability)

- 동시 등장 확률 P(k|i)는 동시 등장 행렬로부터 특정 단어 i의 전체 등장 횟수를 카운트하고, 특정 단어 i가 등장한 횟수를 카운트하여 계산한 조건부 확률(i - 중심 단어, k - 주변 단어)
- P(k|i)에서 i를 중심 단어, k는 주변 단어이며, 동시 등장 행렬에서 중심 단어 i의 행의 모든 값을 더한 값을 분모로 하고 i행 k열의 값을 분자로 한 값이라고 볼 수 있다.

- GloVe 모델은 동시 등장 행렬과 동시 등장 확률을 기반으로 word embedding 수행

```python
import re
from lxml import etree
import urllib.request
import zipfile

from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
with zipfile.ZipFile('ted_ed-20160408.zip', 'r') as z:
    target_text = etree.parse(z.open('ted_en-20160408.xml', 'r'))
    parse_text = '\n'.join(target_text.xpath('//content/text()'))
```

```python
content_text = re.sub(r'\([^)]*\)', '', parse_text)
sent_text = sent_tokenize(content_text)

normalized_text = []
for string in sent_text:
    tokens = re.sub(r'[^a-z0-9]+', ' ', string.lower())
    normalized_text.append(tokens)
    
result = [word_tokenize(sentence) for sentence in normalized_text]
```

```python
from glove import Corpus, Glove
```

```python
corpus = Corpus()
corpus.fit(result, window=5) # 훈련 데이터로부터 동시 등장 행렬 생성
```

### Word Embedding을 Embedding()을 이용하여 직접 구성하거나, 사전 학습된 word embedding

- keras는 훈련 데이터의 단어들에 대한 워드 임베딩을 수행하는 도구인 Embedding()을 제공한다.
- Embedding()은 인공 신경만 구조에서 임베딩 층을 구현한다
    - 임베딩 층의 입력으로 사용하기 위해서는 입력 시퀀스의 각 단어들은 모두 정수 인코딩이 되어 있어야 한다.
    - 임베딩 층은 입력 정수에 대해 밀집 벡터로 맵핑하고 이 밀집 벡터는 인공 신경망의 학습 과정에서 가중치가 학습되는 것과 같은 방식으로 훈련된다.
    - 훈련 과정에서 단어는 모델이 풀고자하는 작업에 맞는 값으로 업데이트 된다.
    - 밀집 벡터를 임베딩 벡터라 한다.

- Embedding(vocab_size, output_dim, input_length)
    - vocab_size : 텍스트 데이터의 전체 단어 집합의 크기
    - input_dim : 워드 임베딩 후의 임베딩 벡터의 차원
    - input_length : 입력 시퀀스 길이
    
- Embedding()은 (number_of_samples, input_length)의 2D 정수 텐서를 입력받는다.
- Embedding()은 (number_of_samples, input_length, embedding_word_dimentionality)의 3D텐서를 반환한다.

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocesiing.sequence import pad_sequences
import numpy as np
```

```python
sentence = ['nice great best amazing', 'stop lies', 'pitiful nerd', 'excellent work', 'supreme quality', 'bad', 'highly respec table']
y_train = [1, 0, 0, 1, 1, 0, 1]
```

```python
t = tokenizer()
t.fit_on_texts(sentence)
vocab_size = len(t.word_index) + 1

print(vocab_size)
```

```python
X_encoded = t.texts_to_sequences(sentence)
print(X_encoded)
```

```python
max_len = max(len(l) for l in X_encoded)
print(max_len)
```

```python
X_train = pad_sequences(X_encoded, maxlen=max_len, padding='pre')
y_train = np.array(y_train)
print(X_train)
print(y_train)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, Flatten
```

```python
model = Sequential()

model.add(Embedding(vocab_size, 4, input_length=max_len))
model.add(Flatten())
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
model.fit(X_train, y_train, epochs=100, verbose=2)
```
